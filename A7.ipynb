```python
import nltk
import warnings
warnings.filterwarnings('ignore')
nltk.download('all')
```

```python
text = ""
with open('testdoc.txt') as f:
    lines = f.readlines()
for line in lines:
    text += line
text = text.replace('\n','')
print(text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
print(sent_tokenize(text))
```

```python
print(word_tokenize(text))
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
wordlist=word_tokenize(text)
```

```python
wordslist=[word for word in wordlist if word not in stop_words]
wordlist
```

```python
tagged=nltk.pos_tag(wordslist)
print(tagged)
```

```python
from nltk.stem.porter import PorterStemmer
stemmed=[PorterStemmer().stem(w) for w in wordlist]
print(stemmed)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmitized=[WordNetLemmatizer().lemmatize(w) for w in wordlist]
print(lemmitized)
```
