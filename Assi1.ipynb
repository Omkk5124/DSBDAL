```python
df["Price of product(available)"]= df["Price of product(available)"].str.replace(',' , '').astype(float)
```

```python
df["price in first site"]= df["price in first site"].str.replace(',' , '').astype(float)
df["price in second site"]= df["price in second site"].str.replace(',' , '').astype(float)
df.dtypes
```

```python
# convert categorical values to numeric
df['first site'].replace(['Amazon', 'Flipkart','Tatacliq'], [0, 1,2], inplace=True)
df['second site'].replace(['Amazon', 'Flipkart','Tatacliq'], [0, 1,2], inplace=True)
df.head(10)
```
