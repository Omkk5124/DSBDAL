```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=sns.load_dataset("titanic")
```

```python
df=pd.DataFrame(titanic)
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df['deck'].fillna(df['deck'].mode()[0],inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(df['age'])
```

```python
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1

low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
```

```python
df=df[(df['age']>=low_lim) & (df['age']<=up_lim)]
df.shape
```

```python
sns.boxplot(df['age'])
```

```python
sns.boxplot(x='sex', y='age', data=df)
```

 The first quartile starts at around 5 and ends at 22 which means that 25% of the passengers are aged between 
5 and 25. The second quartile starts at around 23 and ends at around 32 which means that 25% of the passengers
are aged between 23 and 32. Similarly, the third quartile starts and ends between 34 and 42, hence 
25% passengers are aged within this range and finally the fourth or last quartile starts at 43 and ends around 65.

```python
sns.boxplot(x='sex', y='age', data=df, hue="survived")
```

```python
sns.barplot(x='sex', y='age', data=df)
```

```python
sns.barplot(x='sex', y='age', data=df,hue='survived')
```

```python
sns.violinplot(x='sex', y='age', data=df)
```

```python
sns.violinplot(x='sex', y='age', data=df, hue='survived')
```

if you look at the bottom of the violin plot for the males who survived (left-orange), you can see that 
it is thicker than the bottom of the violin plot for the males who didn't survive (left-blue). 
This means that the number of young male passengers who survived is greater than the number of young male passengers who did not survive.

```python
sns.violinplot(x='sex', y='age', data=df, hue='survived', split=True)
```

```python
sns.stripplot(x='sex',y='age',data=df)
```

```python
sns.stripplot(x='sex',y='age',data=df,hue="survived")
```

you can see there are more points for the males who survived near the bottom 
of the plot compared to those who did not survive.

```python
sns.swarmplot(x='sex',y='age',data=df,hue='survived')
```

```python
sns.swarmplot(x='sex',y='age',data=df,hue='survived',split=True)
```

ratio of surviving males is less than the ratio of surviving females. Since for the male plot,
there are more blue points and less orange points. On the other hand, 
for females, there are more orange points (surviving) than the blue points (not surviving).
